```python
import numpy as np
```

```python
N = len(df_random)
target = 1000
chunk_size = N // target
```

```python
rows_random = []

for i in range(target):
    start_idx = i * chunk_size
    end_idx = start_idx + chunk_size
    chunk = df_random.iloc[start_idx:end_idx]
    rows_random.append(chunk.mean(numeric_only=True))

df_random_reduced = pd.DataFrame(rows_random)
```
