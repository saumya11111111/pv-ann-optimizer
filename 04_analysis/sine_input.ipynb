```python
import pandas as pd
```

```python
df_sine = pd.read_csv('../03_data/PV_results_sine.csv')
```

```python
df_sine.head()
```

```python
df_sine.isnull().sum()
```

```python
df_sine_clean = df_sine[~(df_sine < 0).any(axis=1)].reset_index(drop=True)
```

```python
df_sine_copy = df_sine_clean.copy()
# number of desired samples 
target_rows = 1000
# computing how many rows to combine per avg
rows_per_avg = len(df_sine_copy) // target_rows
print(f"Combining every {rows_per_avg} rows to get approximately {target_rows} samples.")
```

```python
# grouping and averaging
df_sine_reduced = df_sine_copy.groupby(df_sine_copy.index // rows_per_avg).mean()
print(f"New shape after reducing: {df_sine_reduced.shape}")
```

```python
df_sine_reduced.to_csv('../03_data/desampled/PV_results_sine_reduced.csv', index=False)
df_sine_reduced.head
```

```python
flag = (df_sine_reduced < 0).any().any()

if flag:
    print("There are negative values!")
else:
    print("No negative values found.")

```
