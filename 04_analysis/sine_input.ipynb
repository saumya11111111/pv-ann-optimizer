```python
import pandas as pd
```

```python
df_sine = pd.read_csv('../03_data/PV_results_sine.csv')
```

```python
# averaging and downsampling 
num_bins = 1000
bin_size = len(df_sine) // num_bins

df_sine_reduced = df_sine.groupby(df_sine.index // bin_size).mean()
```

```python
df_sine_reduced.to_csv('../03_data/desampled/PV_results_sine_reduced.csv', index=False)
```
